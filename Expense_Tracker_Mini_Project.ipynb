```python
# Expense Tracker Project
expenses_types = []
expenses_cost = []

print('Welcome to Expense Tracker')
while True:
    print('1. Add an expense')
    print('2. Remove an expense')
    print('3. View all expenses')
    print('4. View Total cost')
    print('5. Exit')

    option = input('Please choose an option (1–5): ')

    if option == '1':
        expense_name = input('What type of expense is it? ')
        cost = float(input('What is the cost of the expense? '))
        expenses_types.append(expense_name.strip().lower())
        expenses_cost.append(cost)
        print('Expense added.')

    elif option == '2':
        if not expenses_types:
            print('No expenses to remove.')
        else:
            name_to_remove = input('Enter the expense type to remove: ').strip().lower()
            if name_to_remove in expenses_types:
                idx = expenses_types.index(name_to_remove)
                expenses_types.pop(idx)
                expenses_cost.pop(idx)
                print('Expense removed.')
            else:
                print('Expense not found.')

    elif option == '3':
        if  expenses_types  == []:
            print('No expenses added yet.')
        else:
            print('All expenses:')
            for i in range(len(expenses_types)):
                print(expenses_types[i], '- $', expenses_cost[i])

    elif option == '4':
        print('Total cost of all expenses: $', round(sum(expenses_cost), 2))

    elif option == '5':
        print('Exit successful.')
        break

    else:
        print('Invalid option. Please enter a number from 1 to 5.')
```
