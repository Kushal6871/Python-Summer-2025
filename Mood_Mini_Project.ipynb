```python
# Mood Project
1 == 'Sad'
2 == 'Not great'
3 == 'Okay'
4 == 'Great'
5 == 'Amazing'
import numpy as np
# Average Function
num_list = []
def average(mood_num_list):
  average_mood = np.mean(mood_num_list)
  rounded_average = np.round(average_mood,1)

  if rounded_average >= 4:
    print('Your overall day was wonderful')
  elif rounded_average < 4 and rounded_average > 1:
    print('Your overall Day was okay')
  else:
    print('You had a bad day')
  return rounded_average
print('Your average mood as a number is:',average(num_list))
# Highest Funtion
def highest(mood_num_list):
  highest_mood = np.max(mood_num_list)
  return highest_mood
print('Your highest mood as a number:',highest(num_list))
# Lowest Function
def lowest(mood_num_list):
  lowest_mood = np.min(mood_num_list)
  return lowest_mood
print('Your lowest mood as a number:',lowest(num_list))
# Standard Deviation Fucntion
def standard(mood_num_list):
  standard_deviation = np.std(mood_num_list)
  rounded_standard_deviation = np.round(standard_deviation,2)
  if rounded_standard_deviation >= 2:
    print('Your mood seemed to change a lot throughout the day')
  return rounded_standard_deviation
print('The standard deviation number indicating your change in mood (The higher the more your mood changed throughout your day):',standard(num_list))
```
