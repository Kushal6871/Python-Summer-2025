```python
# Grocery List Project
print('WELCOME TO GROCERY LIST')
groccery_list = []
while True:
  print('1. Add item')
  print('2. Remove item')
  print('3. View List')
  print('4. Exit')
  print('5. Clear List')

  option = int(input('Which option would you like to access, (Type in number of option)'))
  if option == 1:
    choice = str(input('Great, Which item would you like me to add to the list?')).strip().lower()
    groccery_list.append(choice)
    print('Here is your updated Grocery List:',groccery_list)
  elif option == 2 and groccery_list == [] and choice not in groccery_list:
    choice = str(input('Great, Which item would you like me to add to the list?')).strip().lower()
    print('Grocery item is not in list, Please try again')
    print('Here is your updated Grocery List:',groccery_list)
  elif option == 2:
    choice = str(input('Great, Which item would you like me to remove from the list?')).strip().lower()
    groccery_list.remove(choice)
    print('Here is your updated Grocery List:',groccery_list)
  elif option == 3:
    print('Here is your updated Grocery List:',groccery_list)
  elif option == 5:
    print('List Cleared!')
    groccery_list.clear()
  elif option == 4:
    print('Exited')
    break
  else:
    print('Invalid, please try again')


```
