```python
# Dice Roll Simulation Project
import numpy as np
def dice_roll():
  times_rolled = 0
  outcomes = []
  while times_rolled < 15:
    times_rolled += 1
    results = np.random.randint(1,7)
    outcomes.append(results)
  return outcomes
def average_roll(outcomes):
  mean = np.mean(outcomes)
  return round(mean)

def dice_roll_count(outcomes):
  dice_roll_1 = 0
  dice_roll_2 = 0
  dice_roll_3 = 0
  dice_roll_4 = 0
  dice_roll_5 = 0
  dice_roll_6 = 0
  for i in outcomes:
    if i == 1:
      dice_roll_1 += 1
    elif i == 2:
      dice_roll_2 += 1
    elif i == 3:
      dice_roll_3 += 1
    elif i == 4:
      dice_roll_4 += 1
    elif i == 5:
      dice_roll_5 += 1
    elif i == 6:
      dice_roll_6 += 1
    else:
      print('error')
  return dice_roll_1,dice_roll_2,dice_roll_3,dice_roll_4,dice_roll_5,dice_roll_6
rolls = dice_roll()
count_1, count_2, count_3, count_4, count_5, count_6 = dice_roll_count(rolls)
print("Out of the 15 dice rolls:")
print(count_1, "were 1's")
print(count_2, "were 2's")
print(count_3, "were 3's")
print(count_4, "were 4's")
print(count_5, "were 5's")
print(count_6, "were 6's")
print('The outcomes of the 15 dice rolls were',rolls)
print('The average dice roll outcome is',average_roll(rolls))
dice_roll_count(rolls)

```
